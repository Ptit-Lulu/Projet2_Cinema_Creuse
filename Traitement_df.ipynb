```python
import pandas as pd
import numpy as np
link = "https://raw.githubusercontent.com/Ptit-Lulu/Projet2_Cinema_Creuse/donnees_df/resultats_cinema3.csv"
df = pd.read_csv(link, sep=";")

```

```python
map_priorité = { #je crée un dictionnaire de rpiorité pourgarder le meilleur titre possible même si l'idéal manque
    "imdbDisplay": 1,
    "alternative": 2,
    "working": 3,
    "tv": 4,
    "festival": 5,
    "video": 6,
    "dvd": 7,
    "\\N": 8
}

df["priorité"] = df["types"].map(map_priorité)
df_sorted = df.sort_values(["tconst", "priorité"]) #je trie dans l'ordre croissant, car les meilleur titre on un chiffre bas
df_unique_title = df_sorted.drop_duplicates(subset="tconst", keep="first") #je garde le premier titre et donc toujours le meilleur disponible
```

```python
max_genres = df_unique_title["genres"].str.split(",").apply(len).max()
print(max_genres)
```

```python
colonne_explosé = df["genres"].str.split(",", expand=True) #je crée des colonnes mais avec des nom par defaut

nouveaux_noms = []
for i in range(colonne_explosé.shape[1]):
    nouveaux_noms.append("genre_" + str(i+1)) # création des noms correcte

colonne_explosé.columns = nouveaux_noms
```

```python
df_unique_title = df_unique_title.join(colonne_explosé)
```

```python
df_unique_title.info()
```

```python
df_unique_title = df_unique_title.drop(columns=["genres", "original_title", "primaryTitle", "attributes", "types", "ordering", "language", "priorité"  ])
df_unique_title.info()
```

```python
df_unique_title[["primaryName", "primaryProfession","knownForTitles"]] = df_unique_title[["primaryName", "primaryProfession","knownForTitles"]].fillna("Inconnu")
df_unique_title[["tagline", "overview","poster_path"]] = df_unique_title[["tagline", "overview","poster_path"]].fillna("Indisponible")
df_unique_title.info()
```
